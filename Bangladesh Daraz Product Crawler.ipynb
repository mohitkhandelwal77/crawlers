```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from selenium.webdriver.support.ui import WebDriverWait
import selenium.webdriver.support.expected_conditions as EC
from tqdm.notebook import tqdm
import time
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from tqdm import tqdm
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
driver = webdriver.Chrome(executable_path = r'chromedriver.exe',chrome_options = chrome_options)
```

```python
inputlist = pd.read_excel(r"Daraz input list.xlsx")
inputlist
```

```python
inputlist.drop_duplicates(inplace=True)
inputlist.dropna(inplace = True)
asinlist = list(inputlist['url'].unique())
asinlist
```

```python
df = pd.DataFrame(columns=['Product ID','Product Title','Brand','Product URL','Video Count','Aplus Image','Image URL',
                           'Description','Highlights','Average Rating','Rating Count','Specifications','Inside Box'])
```

```python
for val in tqdm(range(0,len(asinlist))):
    
    i=asinlist[val]
    print(i)
    darazurl = i
    driver.get(darazurl)
    print(darazurl)
  
    
    try:
        view_more_button = driver.find_element(By.XPATH, "/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div[2]/button").click()
    except:
        pass
    time.sleep(1)
    soup=BeautifulSoup(driver.page_source,"html.parser")
    productcomplete=soup.find_all("div",class_="pdp-block")
        
    try:
        title=soup.find('div',class_="pdp-product-title")
        product_title = title.find('span', class_='pdp-mod-product-badge-title').text
        print("Product Title:", product_title)
    except AttributeError:
        product_title=""
        print("Product title not found.")
        
    try:
        brand=driver.find_element(By.XPATH,"/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[5]/div/a[1]").text
        print("Brand:",brand)
    except:
        try:
            brand=driver.find_element(By.XPATH,"/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div[1]/div[3]/div[1]/ul/li[1]/div").text
            print("Brand:",brand)
        except:
            brand=""
        
    try:
       # Count the number of video slides
        video_count = len(soup.select('.item-gallery__thumbnail-video'))
        print("Number of Videos:", video_count)
    except:
        video_count =""

    try:   
        # Fetch the product highlights
        highlight=soup.find("div",class_="html-content pdp-product-highlights").text
        print("Product Highlights:",highlight)
    except:
        highlight=""
    try:
      # Fetch the product description
        desc=soup.find("div",class_="html-content detail-content").text
        print("Description:",desc)
    except:
        desc=""
    #fetching aplus image
    try:
        aplus=soup.find("div", class_="html-content detail-content")
        aplus_images = aplus.select('div.html-content img')
        # Extract the src attribute of each img tag and store them in a list
        aplus_image_urls = [img['src'] for img in aplus_images]
        # Print the Aplus image URLs
        print("Aplus Image URLs:")
        for url in aplus_image_urls:
            print(url)
    except:
        aplus_image_urls=""
    #Fetch Images    
    try:
        catimage = soup.find("div", class_="next-slick next-slick-outer next-slick-horizontal")
        image_elements = catimage.find_all("img",class_="pdp-mod-common-image item-gallery__thumbnail-image")
        image_urls = [img['src'] for img in image_elements]
        #changing the pixel
        image_url = [url.replace("100x100", "720x720") for url in image_urls]
        # Print the image URLs
        print("Image URL:")
        for url in image_url:
            print(url)
    except:
        image_url=""
        print("No image found")
        
    try:
    # Wait for the summary_div element to be present
        summary_div =driver.find_element(By.CLASS_NAME, "mod-rating")
        # Extract average rating
        average_rating = summary_div.find_element(By.CLASS_NAME, "score-average").text
        print("Average Rating:", average_rating)

        # Extract rating count
        rating_count = summary_div.find_element(By.CLASS_NAME, "count").text
        print("Rating Count:", rating_count)
    except :
        average_rating = ""
        rating_count = ""
        print("Error while fetching ratings:")
    
   # Fetch Features
    try:
        features = soup.find("div", class_="pdp-general-features")
        features_list = features.find("ul", class_="specification-keys").find_all("li")
        features_dict = {}
        for feature in features_list:
            title = feature.find("span", class_="key-title").text.strip()
            value = feature.find("div", class_="html-content key-value").text.strip()
            features_dict[title] = value
    except:
        features_dict = {}

        
    #Inside Box
    try:
        insidebox=soup.find("div",class_="box-content")
        inside_box=insidebox.find("div",class_="html-content box-content-html").text
        print("Whats inside the box:",inside_box)
    except:
        inside_box=""

    # Append the extracted information as a new row to the DataFrame
    data = {'Product ID':i,'Product Title':product_title,'Brand': brand,'Product URL': darazurl,'Video Count':video_count,
            'Aplus Image':aplus_image_urls,'Image URL':image_url,'Description': desc,'Highlights':highlight,
            'Average Rating': average_rating ,'Rating Count': rating_count,'Specifications':features_dict,
            'Inside Box':inside_box}
    df = df.append(data, ignore_index=True)
    time.sleep(1)

    # Print the extracted shop details
print(df)
print(val)
```
